$B_\lambda$ = $\frac{2hc^2}{\lambda^5}$ $\frac{1}{e^{\frac{hc}{\lambda k_B T}}-1}$ x $10^{-8}$ x sr$^{-1}$

Where all constants are in cgs units, and the $10^{-8}$ comes from converting cm to angstroms.

```python
#calculate planck function in cgs
#divide output by 10^8 because we are using units of flux density with angstroms
def planck(lam_angstroms, A, T):
    lam = lam_angstroms * pow(10, -10) * 100 #convert to cm
    h = 6.626*pow(10, -27) #planck constant
    c = 3*pow(10, 10) #speed of light
    k = 1.38*pow(10, -16) #boltzmann constant
    exp = pow(np.e, h*c/(lam*k*T))
    return A*2*h*pow(c, 2)/pow(lam, 5)*(1/(exp-1)) / 1e8
```

```python
plt.figure(dpi=150)
plt.plot(lam1, flam1)
plt.plot(lam1, planck(lam1, T=5000, A=1.4*pow(10, -20)))
plt.xlabel(r'Wavelength ($\AA$)')
plt.ylabel(r'Flux Density (erg * s^-1 * cm^-2 * $\AA$^-1)')
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(lam2, flam2)
plt.plot(lam2, planck(lam2, T=10000, A=2*pow(10, -16)))
plt.xlabel(r'Wavelength ($\AA$)')
plt.ylabel(r'Flux Density (erg * s^-1 * cm^-2 * $\AA$^-1)')
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(lam3, flam3)
plt.plot(lam3, planck(lam3, T=6500, A=2*pow(10, -19)))
plt.xlabel(r'Wavelength ($\AA$)')
plt.ylabel(r'Flux Density (erg * s^-1 * cm^-2 * $\AA$^-1)')
plt.show()
```
